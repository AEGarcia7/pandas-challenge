```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
school_count=len(school_data_complete["school_name"].unique()) 
student_grade=school_data_complete["grade"]
student_count=len(school_data_complete["student_name"])
diff_budget= school_data_complete["budget"].unique()
total_budget= diff_budget.sum()
reading_av= school_data_complete["reading_score"].mean()
math_av= school_data_complete["math_score"].mean()
reading70= len(school_data_complete.loc[(school_data_complete["reading_score"]>=70)])
reading_pass= reading70/len(school_data_complete["reading_score"])*100
math70= len(school_data_complete.loc[(school_data_complete["math_score"]>=70)])
math_pass= math70/len(school_data_complete["math_score"])*100
overall_pass= len(school_data_complete.loc[(school_data_complete["reading_score"]>=70) & 
                                           (school_data_complete["math_score"]>=70)])/student_count*100

district_summary = pd.DataFrame({"Total Schools": school_count,
             "Total Students": student_count,
             "Total Budget": total_budget,
             "Average Math Score": math_av,
             "Average Reading Score": reading_av,
             "% Passing Math": math_pass,
             "% Passing Reading": reading_pass,
             "% Overall Passing": overall_pass}, index=[0])
district_summary["Total Students"]= district_summary["Total Students"].map("{:,.0f}".format)
district_summary["Total Budget"]= district_summary["Total Budget"].map("${:,.2f}".format)
district_summary
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
school_data_rename = school_data_complete.rename(columns={"student_name":"Student Name",
                                                  "school_name": "School Name",
                                                 "math_score":"Math Score",
                                                 "reading_score":"Reading Score",
                                                 "type":"School Type",
                                                 "size": "Total Students",
                                                 "budget":"Budget",
                                                 "student_grade":"Grade"})

school_mod= school_data_rename[["School Name","School Type","grade","Total Students","Budget","Student Name", "Reading Score" , "Math Score"]]
school_group= school_mod.groupby(["School Name"])
school_group.count()
school_group[["Math Score","Reading Score"]].mean()
av_math=school_group[["Math Score"]].mean()
av_reading=school_group[["Reading Score"]].mean()
math_score=school_mod[(school_mod['Math Score']>=70)] 
reading_score=school_mod[(school_mod['Reading Score']>=70)] 
school_score_overall=school_mod[(school_mod['Reading Score']>=70) & (school_mod['Math Score']>=70)]
school_group=school_mod.groupby(['School Name'])
school_group.head()
school_type=school_group["School Type"].first()
school_budget= school_group["Budget"].first()
total_students=school_group['Total Students'].first()
per_student_budget= school_budget/total_students
reading_score_av=school_group['Reading Score'].mean()
math_score_av= school_group["Math Score"].mean()
math_passers=math_score.groupby(["School Name"]).count()["Math Score"]/school_group["Total Students"].count()*100
reading_passers=reading_score.groupby(["School Name"]).count()["Reading Score"]/school_group["Total Students"].count()*100
overall_passers=school_score_overall.groupby(["School Name"]).count()["Reading Score"]/school_group["Total Students"].count()*100


school_summary=pd.DataFrame({"School Type":school_type,
                             "Total Students":total_students,
                             "Total School Budget":school_budget,
                             "Per Student Budget":per_student_budget,
                             "Average Math Score":math_score_av,
                             "Average Reading Score":reading_score_av,
                             "% Passing Math":math_passers,
                             "% Passing Reading":reading_passers,
                             "% Overall Passing":overall_passers})



school_summary["Total School Budget"]=school_summary["Total School Budget"].map("${:,.2f}".format)
school_summary["Per Student Budget"]=school_summary["Per Student Budget"].map("${:,.2f}".format)

school_summary
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
ninth=school_mod.loc[(school_mod['grade']=="9th")]
ninth_math_score=ninth.groupby(["School Name"]).mean()["Math Score"]
tenth=school_mod.loc[(school_mod['grade']=="10th")]
tenth_math_score=tenth.groupby(["School Name"]).mean()["Math Score"]
eleventh=school_mod.loc[(school_mod['grade']=="11th")]
eleventh_math_score=eleventh.groupby(["School Name"]).mean()["Math Score"]
twelveth=school_mod.loc[(school_mod['grade']=="12th")]
twelveth_math_score=twelveth.groupby(["School Name"]).mean()["Math Score"]

grade_scores_math=pd.DataFrame({"9th":ninth_math_score,
                             "10th":tenth_math_score,
                             "11th":eleventh_math_score,
                             "12th":twelveth_math_score,})
grade_scores_math
```

## Reading Score by Grade 

* Perform the same operations as above for reading scores

```python

ninth_reading_score=ninth.groupby(["School Name"]).mean()["Reading Score"]
tenth_reading_score=tenth.groupby(["School Name"]).mean()["Reading Score"]
eleventh_reading_score=eleventh.groupby(["School Name"]).mean()["Reading Score"]
twelveth_reading_score=twelveth.groupby(["School Name"]).mean()["Reading Score"]

grade_scores_reading=pd.DataFrame({"9th":ninth_reading_score,
                             "10th":tenth_reading_score,
                             "11th":eleventh_reading_score,
                             "12th":twelveth_reading_score,})
grade_scores_reading
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
school_summary_spending=school_summary.drop(["Total Students","Total School Budget","School Type"],axis=1)
school_summary_spending["Per Student Budget"]=school_summary_spending["Per Student Budget"].replace({'\$': '', ',': ''}, regex=True).astype(float)
school_summary_spending.dtypes
spend_bins= [0,585,630,645,680]
spend_bin_names=["<$585","$585-630","$630-645","$645-680"]
school_summary_spending["Spending Ranges (Per Student)"] = pd.cut((school_summary_spending["Per Student Budget"]), spend_bins, labels=spend_bin_names, include_lowest=True)
school_summary_spending=school_summary_spending.drop(["Per Student Budget"],axis=1)
school_summary_spending.groupby("Spending Ranges (Per Student)").max()
```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
school_summary_size=school_summary.drop(["Per Student Budget","Total School Budget","School Type"],axis=1)
size_bins=[0,1000,2000,5000]
size_bin_names=["Small (<1000)","Medium (1000-2000)","Large (2000-5000)"]
school_summary_size["School Size"] = pd.cut((school_summary_size["Total Students"]), size_bins, labels=size_bin_names, include_lowest=True)
school_summary_size=school_summary_size.drop(["Total Students"],axis=1)
school_summary_size.groupby("School Size").max()
```

## Scores by School Type

* Perform the same operations as above, based on school type

```python
school_summary_type=school_summary.drop(["Per Student Budget","Total School Budget","Total Students"],axis=1)
school_summary_type["School Type"]=school_summary_type["School Type"].replace("Charter",0)
school_summary_type["School Type"]=school_summary_type["School Type"].replace("District",2)
type_bins=[0,1,2]
type_bin_names=["Charter","District"]
school_summary_type["School Type"]= pd.cut((school_summary_type["School Type"]), type_bins, labels=type_bin_names, include_lowest=True)
school_summary_type.groupby("School Type").max()
```
